```python
# We use import pandas library because pandas is a powerful Python library for data analysis and manipulation.
#It provides data structures like DataFrame and Series, which are essential for working with structured data.
import pandas as pd

# matplotlib is a popular Python library used for data visualization.
# pyplot is a sub-module of matplotlib that provides a simple interface for creating plots, charts, and graphs.
import matplotlib.pyplot as plt

# seaborn is built on top of matplotlib and makes it easier to create beautiful, informative charts.
# It comes with better default styles and simplified functions for common data visualizations.
import seaborn as sns
```

#### Data Loading into a Dataframe


```python
# We can use pd.read_csv() automatically turns the CSV data into a Pandas DataFrame, which is easy to analyze and manipulate.

data= pd.read_csv("C:\\Users\\Lenovo\\Desktop\\student\\student-mat.csv", sep=";")
print ("Data Load Successfully")
```

#### Check for missing values


```python
# We use .isnull().sum() to check for missing (null/NaN) values in a DataFrame or Series, and count how many are in each column.

data.isnull().sum()
```

#### Display the first few rows using .head()


```python
# We use .head() to quickly view the first few rows of a DataFrame or Series — usually the top 5 rows by default.
# if we want to show more record then we have to mention specific count inside bracket for e.x data.head(10), now it will display top 10 records.

data.head()
```

#### Check data type & missing values


```python
# We use .info() to get a quick summary of a DataFrame & Helps you understand the basic structure of your dataset. 
# for e.x like (No. of rows & Columns, Columns names & data types, No. of non-null values etc.)

data.info()
```

#### Display Column data type using .dtypes


```python
# We use .dtypes to check the data types of each column in a DataFrame or Series.
# Shows whether each column is int, float, object (string), bool, datetime, etc.

print(data.dtypes)
```

#### Understand the dataset's size using .shape.
 

```python
# We use .shape to find the number of rows and columns in a DataFrame or Series.
# It tells you how many rows (records) and columns (features) are in your dataset.
# Helps understand the structure of the dataset before analysis.

print (data.shape)
```

#### Remove duplicate entries using .drop_duplicates()


```python
# We use .drop_duplicates() to remove duplicate rows from a DataFrame, keeping only the first occurrence. 
# Duplicates can skew your analysis or model results.
# Keeps your dataset accurate and reliable.

data.drop_duplicates()
```

#### Data Analyst Questions

```python
# What is the average score in math(G3)?

avg_score= data["G3"].mean()

print(f"Average final math grade (G3): {avg_score}")
```

```python
# How Many student scored above 15 in their final grade (G3)?

stu_above15 = len(data[data["G3"] >15])
print(f"Number of Student scored above 15: {stu_above15}")
```

```python
# Is there a correlation between study time (study time) and the final grade (G3)?

corr= data["studytime"].corr(data["G3"])
print(f"Correlation between Study time & Final grade: {corr}")
```

```python
# Which gender has a higher average final grade (G3)?

avg_gender= data.groupby("sex")["G3"].mean()
print(f"Average final gender grade: {avg_gender}")
```

#### Data Visualization

```python
# Plot a histogram of final grades(G3).

# We use plt.figure(figsize=()) in Matplotlib to set the size of the figure (plot window) in inches before plotting.
# We use bins to show how many block we will display in our visualization,color use to show in which color we represent our visualization,
# edgecolor use to display the colour of edge of bin & linewidth is use to highlight the edge of bin 
# We use .show() to display the final plot window on the screen.


plt.figure(figsize=(20,10))
plt.hist(data["G3"], bins=10, color="Yellow", edgecolor="black", linewidth= 2)
plt.title("Distribution of Final Grades (G3)", size= 25)
plt.xlabel("Final Grade",size= 25)
plt.ylabel("Frequency", size= 25)
plt.show()
```

```python
# Create a scatter plot between study time (study time) and final grade (G3).

plt.figure(figsize=(18,7))
sns.scatterplot(data=data, x="studytime", y="G3", hue="sex")
plt.title("Study time vs Final grade ", size= 25)
plt.xlabel("Study time", size= 25)
plt.ylabel("Final grade", size= 25)
plt.legend(title="Gender")
plt.show()
```

```python
# Create a bar chart comparing the average scores of male and female students.

plt.figure(figsize=(5,3))
data.plot(kind="bar", color=["blue","pink"])
plt.title("Average Final Grade by Gender")
plt.ylabel("Average Final Grade", size=15)
plt.xlabel("Gender", size=20)
plt.xticks(rotation=0)
plt.show()
```
